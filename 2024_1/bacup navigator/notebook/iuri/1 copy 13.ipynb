```python
import bluerobotics_navigator as navigator
from IPython.display import clear_output
import time
import matplotlib.pyplot as plt



# Definir pinos dos motores
PWM_PIN_M = navigator.PwmChannel.Ch4 # Pino PWM para o motor 
IN1_PIN_M = navigator.PwmChannel.Ch3  # Pino IN1 para o motor 
IN2_PIN_M = navigator.PwmChannel.Ch2  # Pino IN2 para o motor 

# Definir pino do servo motor
SERVO_PIN = navigator.PwmChannel.Ch1 # Pino para o servo motor


# Inicializa a navigator 
navigator.init()
navigator.set_pwm_freq_hz(50)


# Funções de controle dos motores
def motor_forward():
    navigator.set_pwm_channel_duty_cycle(IN1_PIN_M, 0)
    navigator.set_pwm_channel_duty_cycle(IN2_PIN_M, 1)
    navigator.set_pwm_channel_duty_cycle(PWM_PIN_M, 1)
    navigator.set_pwm_enable(True)
# Função do servo
def direction_servo(direction=0.105):
    navigator.set_pwm_enable(True)
    if direction == 'direita':
        x = 0.12
    elif direction == 'esquerda':
        x = 0.08
    elif direction == 'reto':
        x = 0.105
    else:
        x = direction
    navigator.set_pwm_channel_duty_cycle(SERVO_PIN, x)

def motor_backward():
    navigator.set_pwm_channel_duty_cycle(IN1_PIN_M, 1)
    navigator.set_pwm_channel_duty_cycle(IN2_PIN_M, 0)
    navigator.set_pwm_channel_duty_cycle(PWM_PIN_M, 1)
    navigator.set_pwm_enable(True)

def stop_motor():
    navigator.set_pwm_enable(False)
    navigator.set_pwm_channel_duty_cycle(IN1_PIN_M, 0)
    navigator.set_pwm_channel_duty_cycle(IN2_PIN_M, 0)
    navigator.set_pwm_channel_duty_cycle(PWM_PIN_M, 0)




direction_servo() # Inicia com servo centralizado

def dir(x):
    navigator.set_pwm_channel_duty_cycle(SERVO_PIN, x)
	


i = 0
a = 0
histotico = [[],[],[],[]]


while True:
    if i == 500:
        # v[i] = v[i-1] +a[i]
        #vel_anterior = 0
        #vel_atual = 0
        #velocidade = [vel_atual, v]
        #aceleracao m/s^2
        acceleration = navigator.read_accel()
        acx = acx*0.95 + 0.05*acceleration.x #direita (negativo)
        acy = acy*0.95 + 0.05*acceleration.y #frente (negativo)
        acz = acz*0.95 + 0.05*acceleration.z #cima (negativo)

        a += 1
        
     
        histotico[0].append(acx)
        histotico[1].append(acy)
        histotico[2].append(acz)
        histotico[3].append(a)
        
        print(f'ac_x: {acx}')
        print(f'ac_y: {acy}')
        print(f'ac_z: {acz}')
        if acy > 0:
            print('frente')
        elif acy < 0:
            print('ré')
        if acx < 0:
            pass
            #print('direita')
        if acz < 0:
            pass
            #print('cima')
        print('')
        print('--------------------------------------')
        print('')
        
        #velocidade angular rad/s
        angular_velocity = navigator.read_gyro()
        girox = angular_velocity.x #direita (negativo) roll
        giroy = angular_velocity.y #frente pitch
        giroz = angular_velocity.z # yaw
        
        print(f'giro_x: {girox}')
        print(f'giro_y: {giroy}')
        print(f'giro_z: {giroz}')
        if girox < 0:
            pass
            #print('inclinado direita')
        if giroy < 0:
            pass
            #print('mirando cima')
        if giroz > 0:
            print('dobrando direita')

        elif giroz < 0:
            print('dobrando esquerda')
        print('')
        print('--------------------------------------')
        print('')


        if a == 1 :
            motor_forward()

        if a == 1001 :
             motor_backward()

        
            

        
        
        
        # #magnetometro µT
        # mag_field = navigator.read_mag()
        # magx = mag_field.x
        # magy = mag_field.y
        # magz = mag_field.z
        clear_output(wait=True)
        i = 0 
        
    i += 1 

    if a == 2000:
        stop_motor()
                 # Nomes para os gráficos
        labels = ["x", "y", "z","a"]
        
        # Criar subplots
        fig, axs = plt.subplots(len(histotico) - 1, 1, figsize=(10, 15))
        
        # Loop para plotar cada lista
        for idx in range(len(histotico) - 1):
            axs[idx].plot(histotico[-1], histotico[idx], marker='o')
            axs[idx].set_title(labels[idx])
            axs[idx].set_xlabel("i")
            axs[idx].set_ylabel(labels[idx])
        
        # Ajustar layout
        plt.tight_layout()
        plt.show()
        break



'''
# Teste de movimentação
try:

    #dir(0.105)
    motor_forward()
    #print('ligando')
    #time.sleep(2)
    if i == 50000000:
        stop_motor()
        print(distancia)
        print(velocidade)
        print(Absoluta_acc)
        print("fim")
    
    acx = acceleration.x #direita (negativo)
    acy = acceleration.y #frente (negativo)
    acz = acceleration.z #cima (negativo)
    Absoluta_acc = (( acx**2 +acy**2 + acz**2) - g)**0.5
    
    if i == 1:
        g = Absoluta_acc
    velocidade = ant_velo + Absoluta_acc
    distancia = distancia + velocidade

    ant_acc = Absoluta_acc
    ant_velo = velocidade
    ant_dist = distancia
    i += 1
    if i%5000 == 0:
        print("oi")
        
        
    #print('girando')
    
    #motor_forward() # Carrinho anda por 1 segundo
    #dir(0.117)

     
    #time.sleep(60)
    #navigator.set_pwm_enable(False)
    
except:
    navigator.set_pwm_enable(False)
    '''
```
