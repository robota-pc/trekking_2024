```python
import bluerobotics_navigator as navigator
from IPython.display import clear_output
import time
import matplotlib.pyplot as plt
import import_ipynb
import calibracao

navigator.init()
i = 0
a = 0


histotico = [[],[],[],[]]

while True:
    if i == 50000:
        # v[i] = v[i-1] +a[i]
        #vel_anterior = 0
        #vel_atual = 0
        #velocidade = [vel_atual, v]
        #aceleracao m/s^2
        acceleration = navigator.read_accel()
        offsetx = calibracao.calibra_x(1000)
        offsety = calibracao.calibra_y(1000)
        offsetz= calibracao.calibra_z(1000)
        acx = acceleration.x - offsetx  #direita (pos)
        acy = acceleration.y - offsety  #frente (pos)
        acz = acceleration.z - offsetz#cima (negativo)

        a += 1
     
        histotico[0].append(acx)
        histotico[1].append(acy)
        histotico[2].append(acz)
        histotico[3].append(a)
        
        print(f'ac_x: {acx}')
        print(f'ac_y: {acy}')
        print(f'ac_z: {acz}')
        if acy > 0:
            print('frente')
        elif acy < 0:
            print('ré')
        if acx < 0:
            pass
            #print('direita')
        if acz < 0:
            pass
            #print('cima')
        print('')
        print('--------------------------------------')
        print('')
        
        #velocidade angular rad/s
        angular_velocity = navigator.read_gyro()
        girox = angular_velocity.x #direita (negativo) roll
        giroy = angular_velocity.y #frente pitch
        giroz = angular_velocity.z # yaw
        
        print(f'giro_x: {girox}')
        print(f'giro_y: {giroy}')
        print(f'giro_z: {giroz}')
        if girox < 0:
            pass
            #print('inclinado direita')
        if giroy < 0:
            pass
            #print('mirando cima')
        if giroz > 0:
            print('dobrando direita')

        elif giroz < 0:
            print('dobrando esquerda')
        print('')
        print('--------------------------------------')
        print('')

        
        
        
        # #magnetometro µT
        # mag_field = navigator.read_mag()
        # magx = mag_field.x
        # magy = mag_field.y
        # magz = mag_field.z
        clear_output(wait=True)
        i = 0 
        
    i += 1 

    if a == 100:
                     # Nomes para os gráficos
            labels = ["x", "y", "z","a"]
            
            # Criar subplots
            fig, axs = plt.subplots(len(histotico) - 1, 1, figsize=(10, 15))
            
            # Loop para plotar cada lista
            for idx in range(len(histotico) - 1):
                axs[idx].plot(histotico[-1], histotico[idx], marker='o')
                axs[idx].set_title(labels[idx])
                axs[idx].set_xlabel("i")
                axs[idx].set_ylabel(labels[idx])
            
            # Ajustar layout
            plt.tight_layout()
            plt.show()
            break
            
```
